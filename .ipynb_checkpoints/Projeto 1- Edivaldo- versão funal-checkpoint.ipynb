# Pergunta norteadora: 
### Como as taxas de desemprego e crescimento econômico foram influenciados pela criação da EFTA e se houve alguma queda nesses indices para os países que saíram da organização  


# **Definição da organização:** 
##### A EFTA (Associação Europeia de Comércio Livre) é um bloco econômico fundada em 4 de Janeiro de 1960 na cidade de Estocolmo, Suécia, pelo Reino Unido, Portugal, Dinamarca, Noruega, Suíça, Áustria e Suécia. Tinha como objetivo a criação de uma área de comércio livre cujo princípio era acabar com impostos aduaneiros entre os estados-membros, o que naturalmente serviu para fomentar as trocas internacionais no espaço desses países.
  #####  Os atuais membros da organização são Islândia, Liechtenstein, Noruega, Suíça.
  ##### Obs: O país Liechtenstein não possui os dados para análise, então será desconsiderado.

# Hipótese e Mecanismo:

   ### Como o principal objetivo da organização é acabar com impostos alfandegários e fomentar trocas entre os países membros, é natural se esperar que isso aumentaria a oferta por emprego, principalmente na área de venda de produtos, já que a demanda aumentaria devido aos preços menores. Isso iria diminuir o desemprego e também agitaria o mercado, causando assim um crescimento econômico na área.

# Análise:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

Dados retirados do site: https://www.gapminder.org/tools/

```python
desemprego = pd.read_excel('aged_15plus_unemployment_rate_percent.xlsx')
crescimento = pd.read_excel('gdp_total_yearly_growth.xlsx')
WB = pd.read_excel('WorldBank.xlsx')
```

```python
Wb_index = WB.set_index("Country")
```

### Os países escolhidos similares para análise são países europeus bem parecidos em tamanho, clima, região e recursos naturais em geral, o que ajudaria a isolar possíveis variaveis que influenciariam nos parametros medidos, ou seja, isso torna a análise mais crível e consistente. 

```python
membros = ["Iceland", "Norway", "Switzerland"]
paises_exmembros = ["Austria","Denmark","Finland","Portugal","Sweden","United Kingdom"]
paises_similares = ["Italy","Germany","France","Spain","Belgium","Netherlands"]
```

```python
Wb_index.loc[paises_similares,"EFTA"] = "não"
Wb_index.loc[membros,"EFTA"] = "sim"
Wb_index.loc[paises_exmembros,"EFTA"] = "ex"
```

```python
Paises_analisados = Wb_index.loc[["Iceland", "Norway", "Switzerland","Austria","Denmark","Finland","Portugal","Sweden","United Kingdom","Italy","Germany","France","Spain","Belgium","Netherlands"]]
Paises_analisados
```

```python
tabela = Paises_analisados.loc[:,"EFTA"]
tabela
```

## Desemprego:

```python
Un_index = desemprego.set_index("country")
Desemprego = Un_index.loc[["Iceland", "Norway", "Switzerland","Austria","Denmark","Finland","Portugal","Sweden","United Kingdom","Italy","Germany","France","Spain","Belgium","Netherlands"]]
Desemprego
```

```python
Periodo = Desemprego.iloc[:,:48:-1]
Periodo.rename(columns=lambda x:str(x), inplace=True)
### Periodo escolhido por ter dados de todos os países e por todos os ex membros já terem deixado a organização.
Periodo_t = Periodo.T
```

```python
#desemprego pelo tempo por país
medias_desemprego = Periodo.mean()
plt.figure(figsize=(30,10))
for pais in Periodo_t.columns:
    plt.scatter(x=Periodo_t.index, y=Periodo_t[pais])
plt.plot(medias_desemprego.index,  medias_desemprego, color="r")
plt.legend()
plt.title("Desemprego pelo tempo por país")
```

```python
# Adicionando uma coluna para saber se o país faz parte da organização.
comparacao = Periodo.join(tabela, how = 'inner',lsuffix ="_desemprego", rsuffix = "_efta")
comparacao
```

```python
comparacao.boxplot(column="2010", by='EFTA')
#Comparação da taxa de desemprego no ano de 2000
```

## Crescimento econômico:

```python
Ce_index = crescimento.set_index("country")
Crescimento = Ce_index.loc[["Iceland", "Norway", "Switzerland","Austria","Denmark","Finland","Portugal","Sweden","United Kingdom","Italy","Germany","France","Spain","Belgium","Netherlands"]]
Crescimento
```

```python
Periodo_c = Crescimento.iloc[:,:130:-1]
Periodo_c.rename(columns=lambda x:str(x), inplace=True)
### Periodo escolhido por tornar possível comparar dados de antes e depois da criação da organização.
Periodo_t2 = Periodo_c.T
```

```python
#crescimento pelo tempo por país
medias_crescimento = Periodo_c.mean()
plt.figure(figsize=(20,10))
for pais in Periodo_t2.columns:
    plt.scatter(x=Periodo_t2.index, y=Periodo_t2[pais])
plt.plot(medias_crescimento.index,  medias_crescimento, color="r")
plt.legend()
plt.title("Crescimento pelo tempo por país")
plt.axvline(x='1960', color= 'red')
# A linha representa o ano de criação da organização
```

```python
# Adicionando uma coluna para saber se o país faz parte da organização.
comparacao_c = Periodo_c.join(tabela, how = "inner",lsuffix ="_crescimento", rsuffix = "_efta")
comparacao_c
```

## Comparação em um gráfico de dispersão

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('Desemprego por crescimento econômico')
    plt.xlabel('Desemprego')
    plt.ylabel('Crescimento econômico')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
# Análise de países pertencentes ao efta
plt.scatter(comparacao.loc[comparacao.EFTA=="sim","2010"], comparacao_c.loc[comparacao_c.EFTA=="sim", "2010"], color="r")
reta(comparacao.loc[comparacao.EFTA=="sim","2010"], comparacao_c.loc[comparacao_c.EFTA=="sim", "2010"],"r")
plt.grid()
```

```python
# Análise de países não pertencentes ao efta
plt.scatter(comparacao.loc[comparacao.EFTA=="não","2010"], comparacao_c.loc[comparacao_c.EFTA=="não", "2010"], color="b")
reta(comparacao.loc[comparacao.EFTA=="não","2010"], comparacao_c.loc[comparacao_c.EFTA=="não", "2010"],"b")
plt.grid()
```

```python
# Análise de países ex pertencentes ao efta
plt.scatter(comparacao.loc[comparacao.EFTA=="ex","2010"], comparacao_c.loc[comparacao_c.EFTA=="ex", "2010"], color="g")
reta(comparacao.loc[comparacao.EFTA=="ex","2010"], comparacao_c.loc[comparacao_c.EFTA=="ex", "2010"], "g")
plt.grid()
```

```python
plt.scatter(comparacao.loc[comparacao.EFTA=="sim","2010"], comparacao_c.loc[comparacao_c.EFTA=="sim", "2010"], color="r")
plt.scatter(comparacao.loc[comparacao.EFTA=="não","2010"], comparacao_c.loc[comparacao_c.EFTA=="não", "2010"], color="b")
plt.scatter(comparacao.loc[comparacao.EFTA=="ex","2010"], comparacao_c.loc[comparacao_c.EFTA=="ex", "2010"], color="g")
plt.grid()
plt.title('Desemprego por crescimento econômico')
plt.xlabel('Desemprego')
plt.ylabel('Crescimento econômico')
```

```python
plt.scatter(comparacao.loc[comparacao.EFTA=="sim","2010"], comparacao_c.loc[comparacao_c.EFTA=="sim", "2010"], color="r")
plt.scatter(comparacao.loc[comparacao.EFTA=="não","2010"], comparacao_c.loc[comparacao_c.EFTA=="não", "2010"], color="b")
plt.scatter(comparacao.loc[comparacao.EFTA=="ex","2010"], comparacao_c.loc[comparacao_c.EFTA=="ex", "2010"], color="g")
reta(comparacao.loc[comparacao.EFTA=="sim","2010"], comparacao_c.loc[comparacao_c.EFTA=="sim", "2010"],"r")
reta(comparacao.loc[comparacao.EFTA=="não","2010"], comparacao_c.loc[comparacao_c.EFTA=="não", "2010"],"b")
reta(comparacao.loc[comparacao.EFTA=="ex","2010"], comparacao_c.loc[comparacao_c.EFTA=="ex", "2010"], "g")
plt.grid()
```

# Conclusão:
 Assim como previsto na hipótese, a participação no grupo econômico trouxe benefícios evidentes aos países e membros. Os países que estavam na EFTA ou que haviam recem saído do grupo no período entre 1995 e 2017 demonstraram taxas de desemprego claramente menores em relação aos outros países, assim como ao analisar o crescimento ecônomico antes e depois da criação da organização, foi evidenciado uma melhora significativa e que se manteu alto por um bom tempo.
 Outro ponto que se confirmou da hipótese é o fato da relação direta do crescimento do PIB e do aumento de empregos no país, como mostrado nos gráficos de dispersão. O único caso que fugiu a regra foi dos países que deixaram participaram do EFTA, mas isso provavelmente se deve ao fato de existirem alguns pontos muito que puxam muito o reta, causando uma distorsão na reta de ajuste. 
 Em uma análise geral, a EFTA beneficiou muito os países envolvidos em sua organização, causando uma estabilidade até mesmo a países que deixaram a organização posteriormente. Isso nos leva a crer que os países que deixaram a organização ou nem se quer entraram nela não o fizeram por causa de seu desempenho, mas talvez por questões diplomáticas ou pela existitência de outras organizações que as interessase mais (como a União Europeia por exemplo).
